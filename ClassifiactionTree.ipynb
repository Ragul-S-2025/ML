```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("/content/Car_Price_Prediction.csv")
data
```

```python

# label encoding
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

data['Make_encoded'] = le.fit_transform(data['Make'])
data['Model_encoded'] = le.fit_transform(data['Model'])
data['Fuel Type_encoded'] = le.fit_transform(data['Fuel Type'])
data['Transmission_encoded'] = le.fit_transform(data['Transmission'])
data
```

```python
x=data[['Make_encoded','Model_encoded','Year',"Fuel Type_encoded","Transmission_encoded"]]
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)


```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
```

```python

y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


mse = mean_squared_error(y_test, y_pred)
mse
r2=r2_score(y_test, y_pred)
r2
```
