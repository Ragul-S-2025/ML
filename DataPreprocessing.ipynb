```python
# Step 1: Remove commas, strip spaces, lowercase
data['Security_deposit'] = data['Security_deposit'].str.replace(',', '').str.strip().str.lower()

# Step 2: Replace 'no' with 0
data['Security_deposit'] = data['Security_deposit'].replace('no', '0')

# Step 3: Lambda conversion for L and Cr
data['Security_deposit'] = data['Security_deposit'].apply(
    lambda x: int(float(x.replace('l', '')) * 100000) if x.endswith('l') else
              int(float(x.replace('cr', '')) * 10000000) if x.endswith('cr') else
              int(float(x))
)

print(data['Security_deposit'])
```

```python
data['Rent_price'] = data['Rent_price'].astype(str).str.replace(',', '').str.strip().str.lower()

data['Rent_price'] = data['Rent_price'].apply(
    lambda x: int(float(x.replace('l', '')) * 100000) if x.endswith('l') else
              int(float(x.replace('cr', '')) * 10000000) if x.endswith('cr') else
              int(float(x))
)

print(data['Rent_price'])
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

data['Property_type'] = le.fit_transform(data['Property_type'])
data['Seller_type'] = le.fit_transform(data['Seller_type'])
data['Status'] = le.fit_transform(data['Status'])
data['BHK'] = le.fit_transform(data['BHK'])
```

```python
data
```

```python
x=data[["Property_type","Seller_type","Area_sqft","Status","Security_deposit","Bathroom","BHK"]]
```

```python
y=data["Rent_price"]
```

```python

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train
```

```python
"""from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)


y_pred = model.predict(x_test)
y_pred"""
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
y_pred
```

```python

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


mse = mean_squared_error(y_test, y_pred)
mse
r2=r2_score(y_test, y_pred)
r2
```
