```python
# Import needed Libraies and datasets

import pandas as pd
data=pd.read_csv("/content/Car_Price_Prediction.csv")
data
```

```python
data.info()
```

```python
# checkingg null values
data.isnull().sum()
```

```python
# One hot encoding for change categical into numerical

data=pd.get_dummies(data,drop_first=True,dtype=int)
data
```

```python
x=data[["Year","Fuel Type_Electric","Transmission_Manual"]]
x
```

```python
y=data[["Price"]]
y
```

```python
# Test and Train test split

import numpy as np
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# using standard scalar for improving accuracy
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
data=sc.fit_transform(data)
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
# model training
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


mse = mean_squared_error(y_test, y_pred)
mse
r2=r2_score(y_test, y_pred)
r2
```
