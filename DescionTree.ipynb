```python
import pandas as pd
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data[["street",	"city",	"statezip",	"country"]].value_counts()
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()

data['date_encoded'] = le.fit_transform(data['date'])
data['street_encoded'] = le.fit_transform(data['street'])
data['city_encoded'] = le.fit_transform(data['city'])
data['statezip_encoded'] = le.fit_transform(data['statezip'])
data['country_encoded'] = le.fit_transform(data['country'])
```

```python

data
```

```python
x=data[[
"bedrooms",
"bathrooms",
"sqft_living",
"sqft_lot",
"floors",
"waterfront",
"sqft_above",
"sqft_basement",
"yr_built",
"yr_renovated",'date_encoded','street_encoded','city_encoded','statezip_encoded','country_encoded']]
y=data[["price"]]
```

```python
x
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

x_train
```

```python
y_train
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error


mse = mean_squared_error(y_test, y_pred)
mse
r2=r2_score(y_test, y_pred)
r2
```
